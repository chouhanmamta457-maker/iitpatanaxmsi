```python
import pandas as pd
import random

# Set seed for reproducibility
random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print("--- First 10 Rows ---")
print(df.head(10))
print(f"\nDataset shape: {df.shape}")
```

Task 1: Basic Grouping and Single Aggregations

```python
# 1. Total sales per region
regional_sales = df.groupby('region')['sales_amount'].sum().reset_index()
regional_sales_sorted = regional_sales.sort_values(by='sales_amount', ascending=False)

# 2. Number of transactions per category
category_counts = df.groupby('category')['transaction_id'].count().reset_index()

# 3. Average sales per salesperson
salesperson_avg = df.groupby('salesperson')['sales_amount'].mean().reset_index()

print("Top Performing Regions:\n", regional_sales_sorted)
print("\nNumber of Transactions per Category:\n", category_counts)
print("\nAverage Sales per Salesperson:\n", salesperson_avg)
```

Task 2: Multi-Column Grouping and Multiple Aggregations

```python
# 1. Group by Region and Category
region_cat_sales = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()

# 2. Multiple metrics for Salespersons
salesperson_metrics = df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()

# 3. Sort to find top performer
top_salesperson = salesperson_metrics.sort_values(by='sum', ascending=False)

# 4. Find category with maximum total revenue
# We first group by category, then find the index of the max sum
category_revenue = df.groupby('category')['sales_amount'].sum()
top_category = category_revenue.idxmax()

print(f"The top-performing category is: {top_category}")
print("\nSalesperson Performance Metrics:\n", top_salesperson)
```

Task 3: Custom Aggregation and Complete Sales Report

```python
# 1. Define custom function
def sales_range(x):
    return x.max() - x.min()

# 2. Apply custom and standard functions
regional_analysis = df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()

# 3. Dictionary-based aggregation for the Final Report
final_report = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()

print("Final Regional Summary Report:")
print(final_report)
```
