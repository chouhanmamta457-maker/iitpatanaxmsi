```python
import numpy as np
```

```python
data = np.array([[22.5, 19.0, 31.2, 28.7, 25.1],
                 [17.3, 22.8, 30.5, 26.4, 21.9],
                 [33.1, 29.6, 18.4, 24.0, 27.8],
                 [20.2, 23.5, 31.9, 28.1, 22.6]])
```

```python
print("Task 1:")
print(f"Shape of data: {data.shape}")

mean_per_station = data.mean(axis=1)
print(f"Mean temperature per station: {mean_per_station}")
print("-" * 30)
```

```python
print("Task 2:")
high_temps = data[data > 28.0]
print(f"Readings above 28.0°C: {high_temps}")
print("-" * 30)
```

```python
print("Task 3:")
d_min = data.min()
d_max = data.max()

normalized = (data - d_min) / (d_max - d_min)

print("Normalized Data (rounded to 2 decimal places):")
print(np.round(normalized, 2))
```
