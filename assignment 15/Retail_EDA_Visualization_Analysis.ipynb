```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio

np.random.seed(0)
n = 200

df = pd.DataFrame({
    'age':      np.random.randint(18, 60, n),
    'income':   np.random.randint(20000, 120000, n),
    'spend':    np.random.randint(500, 8000, n),
    'category': np.random.choice(['Electronics', 'Clothing', 'Groceries', 'Sports'], n)
})
```

```python

avg_spend = df.groupby('category')['spend'].mean().sort_values()

plt.figure(figsize=(8, 5))
plt.bar(avg_spend.index, avg_spend.values, color='skyblue', edgecolor='black')

plt.title('Average Spend per Category', fontsize=14)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Mean Spend ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='income', y='spend', alpha=0.6)

plt.title('Relationship Between Income and Spend', fontsize=14)
plt.xlabel('Annual Income ($)', fontsize=12)
plt.ylabel('Annual Spend ($)', fontsize=12)

plt.show()
```

```python
pio.renderers.default = 'browser'
fig = px.scatter(df, x='income', y='spend', color='category',
                 title='Interactive Analysis: Income vs Spend by Category',
                 labels={'income': 'Annual Income ($)', 'spend': 'Annual Spend ($)'},
                 hover_data=['age'])

fig.show()
```
