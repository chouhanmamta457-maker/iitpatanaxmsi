```python
import requests

url = "https://restcountries.com/v3.1/all"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

try:
    # Adding headers can often bypass a 400 error
    response = requests.get(url, headers=headers, timeout=10)
    print(f"HTTP Status Code: {response.status_code}")
    
    if response.status_code == 200:
        print(f"Success! Retrieved {len(response.json())} countries.")
    else:
        print(f"Server returned an error: {response.text}")

except Exception as e:
    print(f"Connection error: {e}")
```

```python
# 1. We use .strip() to ensure no accidental trailing spaces are present
url = "https://restcountries.com/v3.1/name/india".strip()

# 2. Add a User-Agent header to mimic a standard browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

try:
    # 3. Pass the headers into the request
    response = requests.get(url, headers=headers, timeout=10)

    response.raise_for_status()

    print(f"HTTP Status Code: {response.status_code}")
    countries_data = response.json()
    print(f"Success! Total number of countries retrieved: {len(countries_data)}")

except requests.exceptions.HTTPError as http_err:
    print(f"HTTP error occurred: {http_err}")
    # If it's still a 400, the server might be having a bad day or rejecting the specific path
except Exception as err:
    print(f"An unexpected error occurred: {err}")
```

```python
import pandas as pd

# Extracting specific fields into a list of dictionaries
extracted_list = []

for country in countries_data:
    country_info = {
        "common_name": country.get("name", {}).get("common"),
        "population": country.get("population", 0),
        "region": country.get("region", "N/A"),
        "area": country.get("area", 0)
    }
    extracted_list.append(country_info)

# Create DataFrame
df = pd.DataFrame(extracted_list)

# Display the first 10 rows
print("First 10 Rows of the Country Dataset:")
df.head(10)
```

```python
# 1. Which region has the highest total population?
region_population = df.groupby("region")["population"].sum()
highest_pop_region = region_population.idxmax()
total_pop = region_population.max()

# 2. Which country has the largest area?
# We use idxmax to find the index of the row with the maximum area
largest_country_idx = df["area"].idxmax()
largest_country_name = df.loc[largest_country_idx, "common_name"]
largest_area_value = df.loc[largest_country_idx, "area"]

print(f"--- Summary Report ---")
print(f"Region with highest total population: {highest_pop_region} ({total_pop:,} people)")
print(f"Country with the largest area: {largest_country_name} ({largest_area_value:,} sq km)")
```
