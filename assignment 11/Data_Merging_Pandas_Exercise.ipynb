```python
import pandas as pd
import numpy as np
```

```python
# === TASK 1: DATA PREPARATION ===
print("=== TASK 1: Data Preparation ===")

# 1. Create DataFrames
students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}

enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}

scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}

df_students = pd.DataFrame(students_data)
df_enrollments = pd.DataFrame(enrollments_data)
df_scores = pd.DataFrame(scores_data)

# 2. Null Value Analysis
null_counts = df_students.isnull().sum()
null_pct = (df_students.isnull().sum() / len(df_students)) * 100

print("\nNull Value Analysis:")
for col in df_students.columns:
    print(f"Column: {col}, Nulls: {null_counts[col]} ({null_pct[col]:.2f}%)")

# 3. Data Cleaning
df_students['city'] = df_students['city'].fillna('Unknown')
df_students = df_students.dropna(subset=['name'])

print("\nCleaned Students DataFrame:")
print(df_students)
```

```python
# === TASK 2: JOIN OPERATIONS ===
print("\n=== TASK 2: Join Operations ===")

# A. Inner Join
inner_join = pd.merge(df_students, df_enrollments, on='student_id', how='inner')
print(f"Inner Join Result ({len(inner_join)} rows)")
# Excluded: 104, 106, 107 because they exist in Students but not in Enrollments.

# B. Left Join
left_join = pd.merge(df_students, df_enrollments, on='student_id', how='left')
print(f"Left Join Result ({len(left_join)} rows)")
# Null course_names: 104, 106, 107 (Students who haven't enrolled in courses yet).

# C. Right Join
right_join = pd.merge(df_students, df_enrollments, on='student_id', how='right')
print(f"Right Join Result ({len(right_join)} rows)")
# student_ids 108, 109 appear without names (They are in Enrollments but not in our Students master list).

# D. Full Outer Join
outer_join = pd.merge(df_students, df_enrollments, on='student_id', how='outer', indicator=True)
print(f"Full Outer Join Result ({len(outer_join)} rows)")
print("\nMerge Source Distribution:")
print(outer_join['_merge'].value_counts())
```

```python
# === TASK 3: LOOKUP AND AUTOMATION ===
print("\n=== TASK 3: Lookup and Automation ===")

# 1. Lookup Operation using .map()
score_map = dict(zip(df_scores['student_id'], df_scores['exam_score']))
df_students['exam_score_map'] = df_students['student_id'].map(score_map)

# 2. Efficiency Note: 
# .map() is generally faster for single-column lookups because it avoids the overhead 
# of aligning multiple columns and managing indices that pd.merge() requires.

# 3. Automation Function
def auto_merge(df1, df2, join_type, key_column):
    merged = pd.merge(df1, df2, on=key_column, how=join_type)
    return {
        'result_df': merged,
        'row_count': len(merged),
        'join_type': join_type
    }

# Testing automation
test_inner = auto_merge(df_students, df_enrollments, 'inner', 'student_id')
print(f"Automation Test (Inner): {test_inner['row_count']} rows found.")
```
